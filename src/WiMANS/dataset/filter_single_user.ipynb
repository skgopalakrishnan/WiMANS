```python
def filter_records(df: pd.DataFrame, user_name:int, pref_location:str, total_users:int = 6):
    """
    Function to extract all the records of single user (user_name) along with the position they are situated at

    Args:
        df: pd.DataFrame, dataframe containing the annotations
        user_name: int, ID of the user
        pref_location: str, preferred location of the user
        total_users: int, total number of users in the dataset
    Returns:
        fil_df: pd.DataFrame, dataframe containing the filtered records
    """

    if user_name == 0:
        uname = ["user_" + str(ii) + str("_location") for ii in range(1, total_users+1)]
    else:
        uname = "user_" + str(user_name) + "_location"
        user_df = df[df["number_of_users"] == 1]
        print("Number of single user records is:", len(user_df))

    if pref_location is None:
        pref_location = ["a", "b", "c", "d", "e"]
    else:
        pref_location = [pref_location]

    # user_locations = ["user_" + str(ii) + str("_location") for ii in range(1, total_users+1)]
    # user_activities = ["user_" + str(ii) + str("_activity") for ii in range(1, total_users+1)]

    fil_df = single_user_df.loc[single_user_df[uname].isin(pref_location)]

    print(f"Number of user records of {user_name} at {pref_location} after filtering is {len(fil_df)}")

    return fil_df
```
